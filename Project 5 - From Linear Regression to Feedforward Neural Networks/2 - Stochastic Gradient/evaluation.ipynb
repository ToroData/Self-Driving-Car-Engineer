# Evaluation of accuracy

In this section, we will evaluate the accuracy of the trained model by visualizing the accuracy history over epochs. The accuracy history represents the performance of the model on the training dataset at each epoch during the training process.

We will use Matplotlib and Plotly to create visualizations of the accuracy history.

First, let's import the necessary libraries:

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
pio.renderers.default = 'png'
import plotly.io as pio
import numpy as np
import subprocess
import json
```

## Training

Next, we will run the training script to train the model and store the accuracy history. The training script (training.py) takes the image directory (GTSRB/Final_Training/Images/) and the number of epochs (50 in this case) as input parameters.

```python
output = subprocess.check_output(['python', 'training.py', '--imdir', 'GTSRB/Final_Training/Images/', '--epochs', '50'])
```

## Visualizing epochs

Now, we can load the accuracy history from the generated JSON file (accuracy_history.json) and create a list of epochs.

```python
with open('accuracy_history.json', 'r') as f:
    accuracy_history = json.load(f)
```

We will use Plotly to create an interactive line plot to visualize the accuracy history over epochs. The x-axis represents the epochs, and the y-axis represents the accuracy.

```python
epochs = list(range(len(accuracy_history)))

fig = go.Figure()
fig.add_trace(go.Scatter(x=epochs, y=accuracy_history, mode='lines', name='Train Accuracy'))

fig.update_layout(
    title='Accuracy vs. Epochs',
    xaxis_title='Epochs',
    yaxis_title='Accuracy',
    legend=dict(
        x=0.7,
        y=1,
        traceorder='normal',
        font=dict(
            family='sans-serif',
            size=12,
            color='black'
        ),
        bgcolor='white',
        bordercolor='gray',
        borderwidth=1
    )
)
fig.write_html('accuracy_graph.html')
```

The code above creates an interactive plot that can be displayed directly in the Jupyter Notebook. Additionally, I save the plot as an HTML file (accuracy_graph.html) for future reference.

Finally, let's display the plot:

```python
fig.show()
```

The resulting plot will show the accuracy of the model on the training dataset at each epoch. This visualization can help us understand how the accuracy improves or converges over the training process.
